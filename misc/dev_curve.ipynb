# Cap

```python
c = Cap()
c.build()
```

```python
plt.scatter(c.x,c.y,s=1,c=c.a)
```

```python
def diff(a,b):
    return a[1:], np.diff(b)/np.diff(a)
```

```python
plt.plot(c.a,c.dx_da)
plt.plot(*diff(c.a,c.x))
```

```python
plt.plot(c.a,c.ddx_dda)
plt.plot(*diff(c.a,c.dx_da))
```

```python
plt.plot(c.a,c.dy_da)
plt.plot(*diff(c.a,c.y))
```

```python
plt.plot(c.a,c.ddy_dda)
plt.plot(*diff(c.a,c.dy_da))
```
