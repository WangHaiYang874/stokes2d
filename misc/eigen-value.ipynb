```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(4.21225,4.21250,400)[1:]
y = np.linspace(2.2505,2.2510,400)[1:]
x,y = np.meshgrid(x,y,sparse=1)
z = x + 1j*y
```

```python
fig, ax = plt.subplots(figsize=(20,20))
val = np.abs(np.sin(z) + z)
ax.scatter(z.real,z.imag,c=val,s=1)

xi,yi = np.unravel_index(np.argmin(np.abs(np.sin(z) + z)),z.shape)
z_argmin = z[xi,yi]
im = ax.scatter(z_argmin.real,z_argmin.imag,c='r',s=3)

ax.set_aspect(1)
ax.grid()
fig.colorbar(im)
print('the red dot has value of ', val[xi,yi])
print('the real part of the red dot is', z_argmin.real)
```
