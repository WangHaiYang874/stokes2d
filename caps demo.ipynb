```python
from caps import smooth_cap
from matplotlib import pyplot as plt
import numpy as np
```

## questions

- the curvature looks pretty weird. 

## Demonstrations

```python
cap = smooth_cap()
plt.plot(cap.x,cap.y)
plt.title('the standard cap')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

I can get the 
- parameter `a`, 
- integral weight `da`
- points on the curve `t`
- derivative of the parametrization `dt_da`
- curvature (signed) `k`

with the following command

```python
a,da,t,dt_da, k = cap.get_data()
```

And this is the plot for the derivative

```python
plt.plot(cap.a,cap.dx_da,label='dx/da')
plt.plot(cap.a,cap.dy_da,label='dy/da')
plt.plot(cap.a,cap.get_k(),label='curvature')
plt.legend()
plt.title('the derivatives of the standard cap')
plt.xlabel('parameter a')
plt.show()
```

## rotated and scaled cap
