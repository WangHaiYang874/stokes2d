```python
from matplotlib import pyplot as plt
import numpy as np
from io import BytesIO
from imageio import imread
```

```python
x = np.linspace(0,5,5)
y = x + 1

x, y = np.meshgrid(x, y)
z = x**2 + y**2

plt.imshow(z, extent=(0,5,0,6), origin='lower', interpolation='gaussian', vmin = 0, vmax = 50)

layer = np.zeros_like(z)
layer[3,3] = 100000

masked = np.ma.masked_where(layer==0, layer)
plt.imshow(masked, extent=(0,5,1,6), origin='lower',  vmin = 0, vmax = 50)
```

```python
fig, ax = plt.subplots()

x = np.linspace(0,5,5)
y = x + 1

x, y = np.meshgrid(x, y)
z = x**2 + y**2

masked = np.ma.masked_where(z<10, z)

ax.imshow(masked, extent=(0,5,0,6), origin='lower', interpolation='gaussian', vmin = 0, vmax = 50)
ax.set_axis_off()
```

```python
byte_file = BytesIO()
fig.savefig(byte_file, format='png')
byte_file.seek(0)
```

```python
img = imread(byte_file)
```
